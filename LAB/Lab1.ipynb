## 1. MÔ HÌNH HỒI QUY TUYẾN TÍNH CƠ BẢN

Khởi tạo dữ liệu: khởi tạo 100 điểm dữ liệu gồm X và y.

$$y = 4 + 3*X+  \varepsilon$$   

```python
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100,1)
```

Thêm giá trị bias_term vào dữ liệu X ban đầu. Mặc định bias_term là 1.  

```python
X_b = np.c_[np.ones((100, 1)), X] 
```

### Câu 1. Vẽ biểu đồ cho dữ liệu (X,y) vừa tạo ở trên.  

```python
import matplotlib.pyplot as plt
plt.plot(X, y, 'b.')
plt.show()
```

### Câu 2. Cho biết chiều (shape) của X_b và y.  


```python
X_b.shape
```

### Câu 3. Hãy viết code Python tìm ra tham số tối ưu cho mô hình hồi quy tuyến tính. Tìm tham số tối ưu cho mô hình bằng phương trình sau: 

$$\hat{\theta} = (X^TX)^{-1}X^Ty$$

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

Tạo ra dữ liệu mới X_new bằng đoạn lệnh sau (dùng để kiểm tra):

```python
X_new = np.array([[0], [2]]) 
X_new_b = np.c_[np.ones((2, 1)), X_new] 
```

### Câu 4. Dùng tham số tối ưu của hồi quy tuyến tính vừa tìm được ở Câu 3, dự đoán kết quả cho X_new_b. Giá trị dự đoán lưu vào biến y_pred. 

$$\hat{y} = X\hat{\theta}$$

```python
y_pred = X_new_b.dot(theta_best)
y_pred
```

### Câu 5. Vẽ biểu đồ thể hiện kết quả dự đoán trên dữ liệu. 

```python
import matplotlib.pyplot as plt
plt.title('Biểu đồ thể hiện kết quả dự đoán')
plt.plot(X,y,'b.')
plt.plot(X_new, y_pred, 'r-')
plt.show()
```

# 2. BÀI TẬP

## Bài tập 2. Thực hiện xây dựng mô hình hồi quy đa thức (Polynomial Regression) cho dữ liệu đa thức như sau: 

Dữ liệu huấn luyện:  

```python
import numpy as np
m = 100 
X = 6 * np.random.rand(m, 1) - 3 
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1) 
```

Dữ liệu dự đoán: 

```python
X_new = np.linspace(-3, 3, 100).reshape(100, 1) 
```

Sử dụng lớp PolynomialFeatures trong sklearn để chuyển giá trị trong X và X_new về thành dạng bậc 2 (dữ liệu phi tuyến) theo đoạn code mẫu như sau: 

```python
from sklearn.preprocessing import PolynomialFeatures 
poly_features = PolynomialFeatures(degree=2, include_bias=False) 
X_poly = poly_features.fit_transform(X) 
```

### Câu 1. Vẽ biểu đồ cho dữ liệu (X,y) vừa tạo ở trên.  

```python
plt.plot(X, y, 'b.')
plt.show()
```

### Câu 2. Cho biết chiều (shape) của X_b và y.  

```python
X_poly_b = np.c_[np.ones((100, 1)), X_poly] 
```

```python
X_poly_b.shape
```

### Câu 3. Hãy viết code Python tìm ra tham số tối ưu cho mô hình hồi quy tuyến tính. Tìm tham số tối ưu cho mô hình bằng phương trình sau:

```python
theta_best = np.linalg.inv((X_poly_b.T.dot(X_poly_b)))\
            .dot(X_poly_b.T).dot(y)
theta_best
```

### Câu 4. Dùng tham số tối ưu của hồi quy tuyến tính vừa tìm được ở Câu 3, dự đoán kết quả cho X_new_b. Giá trị dự đoán lưu vào biến y_pred. 

```python
X_poly_new = poly_features.fit_transform(X_new)
X_poly_new_b = np.c_[np.ones((100, 1)), X_poly_new]
```

```python
y_poly_pred = X_poly_new_b.dot(theta_best)
y_poly_pred
```

### Câu 5. Vẽ biểu đồ thể hiện kết quả dự đoán trên dữ liệu. 

```python
import matplotlib.pyplot as plt
plt.plot(X, y, 'b.')
plt.plot(X_new, y_poly_pred, 'r-')
plt.show()
```
